# Image to Pencil Sketch with Python:


## About the project


We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

# Import Opencv and Matplotlib

```python
import cv2
import matplotlib.pyplot as plt
```

# read and display the image using cv2

```python
image = cv2.imread("lion.jpg")
img = cv2.resize(image, (960, 560))
cv2.imshow("lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

# convert simple image into grayscale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img = cv2.resize(gray_image, (960, 560))
cv2.imshow("Gray lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# invert the grayscale image : negative image

```python
inver_image = cv2.bitwise_not(gray_image)
img = cv2.resize(inver_image, (960, 560))
cv2.imshow("Inverted Gray lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred = cv2.GaussianBlur(inver_image, (111, 111), 0)
img = cv2.resize(blurred, (960, 560))
cv2.imshow("Blurry Inverted Gray lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# inverted blurry image

```python
invblr = cv2.bitwise_not(blurred)
img = cv2.resize(invblr, (960, 560))
cv2.imshow("Inverted blurry lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# pencil skatch

```python
pencil_sketch = cv2.divide(gray_image, invblr, scale=256.0)
img = cv2.resize(pencil_sketch, (960, 560))
cv2.imshow("Pencil scatch lion", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# comparing both orignal and pencil skatch image side by side

```python
img = cv2.resize(image, (960, 560))
cv2.imshow("lion", img)


temp_img = cv2.resize(pencil_sketch, (960, 560))
cv2.imshow("pencil sketch", temp_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# display final image using matplotlib

```python
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
